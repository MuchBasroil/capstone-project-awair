```python
# Fungsi untuk mengambil data dari API
def fetch_data(latitude, longitude):
    payload = {
        "location": {
            "latitude": latitude,
            "longitude": longitude
        },
        "universalAqi": True,
        "languageCode": "en",
        "extraComputations": [
            "HEALTH_RECOMMENDATIONS",
            "DOMINANT_POLLUTANT_CONCENTRATION",
            "POLLUTANT_CONCENTRATION",
            "LOCAL_AQI",
            "POLLUTANT_ADDITIONAL_INFO"
        ]
    }
    response = requests.post(url, json=payload)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch data for {latitude}, {longitude}. Status code: {response.status_code}")
        return None
```

```python
# Inisialisasi DataFrame untuk menyimpan data
data = []

# Loop untuk mengambil data dari API untuk setiap lokasi
for lat, lon in locations:
    print(f"Fetching data for {lat}, {lon}...")
    result = fetch_data(lat, lon)
    if result:
        try:
            # Ambil data polutan dan rekomendasi kesehatan
            record = {
                "latitude": lat,
                "longitude": lon,
                "dateTime": result["dateTime"],
                "regionCode": result.get("regionCode", "N/A"),
                "universal_aqi": result["indexes"][0]["aqi"],
                "universal_aqi_category": result["indexes"][0]["category"],
                "dominant_pollutant": result["indexes"][0]["dominantPollutant"],
                "co": result["pollutants"][0]["concentration"]["value"],
                "no2": result["pollutants"][1]["concentration"]["value"],
                "o3": result["pollutants"][2]["concentration"]["value"],
                "pm10": result["pollutants"][3]["concentration"]["value"],
                "pm25": result["pollutants"][4]["concentration"]["value"],
                "so2": result["pollutants"][5]["concentration"]["value"],
                "health_general_population": result["healthRecommendations"]["generalPopulation"],
                "health_elderly": result["healthRecommendations"]["elderly"],
                "health_lung_disease": result["healthRecommendations"]["lungDiseasePopulation"],
                "health_heart_disease": result["healthRecommendations"]["heartDiseasePopulation"],
                "health_athletes": result["healthRecommendations"]["athletes"],
                "health_pregnant_women": result["healthRecommendations"]["pregnantWomen"],
                "health_children": result["healthRecommendations"]["children"],
            }
            data.append(record)
        except KeyError as e:
            print(f"KeyError: {e}, skipping this location.")
```
